```python
data="Ml is a subset of Ai which learns the pattern and predict the feature"
chunks=[]
fixed_size=20
for i in range(0,len(data),fixed_size):   #0 20
    chunk = data[i:i+fixed_size]    #data[0:0+20] data[0:20]  data[20:20+20]
    chunks.append(chunk)
print(chunks)
```

```python
data="Ml is a subset of Ai which learns the pattern and predict the feature"
chunks=[]
fixed_size=20
overlapping=4
for i in range(0,len(data),fixed_size-overlapping):   
    chunk = data[i:i+fixed_size]    
    chunks.append(chunk)
print(chunks)
```
