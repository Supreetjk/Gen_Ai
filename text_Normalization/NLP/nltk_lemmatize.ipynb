```python
Lemmatizer=WordNetLemmatizer()
```

create a list of 10 tokens for WordNetLemmatizer 

```python
for word in tokens:
    print(word,Lemmatizer.lemmatize(word,pos='v'))
```

```python
for word in tokens:
    print(word,Lemmatizer.lemmatize(word,pos='n'))    #by default it uses n(noun)
```

'''
pos : str
The Part Of Speech tag. Valid options are 
"n" for nouns, 
"v" for verbs, 
"a" for adjectives, 
"r" for adverbs, 
"s" for satellite adjectives
'''

```python
nltk.pos_tag(tokens)
'''
pos : str
The Part Of Speech tag. Valid options are 
"n" for nouns, 
"v" for verbs, 
"a" for adjectives, 
"r","j" for adverbs, 
"s" for satellite adjectives
'''
```

```python
def find_postag(tag):
    if tag.startswith('J'):
        return 'a'
    elif tag.startswith('N'):
        return 'n'
    elif tag.startswith('V'):
        return 'v'
    elif tag.startswith('R'):
        return 'r'
    else:
        return 'n'
```

```python
pos_tag_words=nltk.pos_tag(tokens)
pos_tag_words
```

```python
for word,tag in pos_tag_words:   #('houses','NNS')
    tagged_word=Lemmatizer.lemmatize(word,find_postag(tag))
    # Lemmatizer.lemmatize('houses','n')
    print(tagged_word)
```
