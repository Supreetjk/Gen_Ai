```python
import nltk
```

```python
nltk.download("punkt")
nltk.download("punkt_tab")
```

```python
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
res='''Return a tokenized copy of text, using NLTK's recommended word tokenizer (currently an improved .TreebankWordTokenizer along with .PunktSentenceTokenizer for the specified language'''
res.split(".")
```

```python
word_tokens = word_tokenize(res)
word_tokens
```

```python
sent_tokens = sent_tokenize(res)
sent_tokens
```

```python
import re

re.findall(r'\s+',res)
# re.sub(r'\s+','',res)
```

```python
re.sub(r'\s{2,}','',res)
```

```python
re.sub(r'[^a-zA-Z0-9]',' ',res)
```
